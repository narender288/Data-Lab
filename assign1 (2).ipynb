```python
# for transforing the dataframe into 2D sparse matrix .
from scipy.sparse import csr_matrix
sparse_matrix00 = csr_matrix(df1)
print(sparse_matrix00)
sparse_matrix00.ndim   # also 2.
sparse_matrix00.dtype  #float64
print(type(sparse_matrix00))
```

```python
n_m = sparse_matrix00.toarray()
print(n_m)
print(n_m.ndim)
print(n_m.shape)
print(type(n_m))
```

```python
import numpy as np
from scipy.sparse import csr_matrix

mat_s = 1000

min_x = np.min(n_m[:, 0])
min_y = np.min(n_m[:, 1])
max_x = np.max(n_m[:, 0])
max_y = np.max(n_m[:, 1])

# we defined min and max as it is need in discretization.

#IMPORTANT as formula i.e[ normalization * (n-1)]
discretized_x = np.clip(((n_m[:, 0] - min_x) / (max_x - min_x) * (mat_s - 1)).astype(int), 0, mat_s - 1)
discretized_y = np.clip(((n_m[:, 1] - min_y) / (max_y - min_y) * (mat_s - 1)).astype(int), 0, mat_s - 1)

print(discretized_x.ndim)  # 1d vector
print(discretized_x.size)  # 801 element in x and 801 in y.

bool_m = np.zeros((mat_s, mat_s), dtype=bool)  # boolean given.

for x, y in zip(discretized_x, discretized_y):                  #zip is simply used to write two tuple,list,vector ect together.
    bool_m[x, y] = True   # discretization important for this step only.
        
sparse_matrix01 = csr_matrix(bool_m)
print(sparse_matrix01)
print(sparse_matrix01.ndim)
print(type(sparse_matrix01))
print(sparse_matrix01.shape)
```

```python
# convert sparsh_matrix to numpymatrix

numpy_1 = sparse_matrix01.toarray()
print(numpy_1)
print(numpy_1.ndim)
print(numpy_1.size)
numpy_1[999,472]
```
