```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.base import BaseEstimator, ClassifierMixin

class RDA(BaseEstimator, ClassifierMixin):
    def __init__(self, lambda_param=0.5):
        self.lambda_param = lambda_param
        self.lda = LDA()
        self.qda = QDA()

    def fit(self, X, y):
        self.lda.fit(X, y)
        self.qda.fit(X, y)
        return self

    def predict(self, X):
        lda_probs = self.lda.predict_proba(X)
        qda_probs = self.qda.predict_proba(X)
        combined_probs = (1 - self.lambda_param) * lda_probs + self.lambda_param * qda_probs
        return np.argmax(combined_probs, axis=1)

    def predict_proba(self, X):
        lda_probs = self.lda.predict_proba(X)
        qda_probs = self.qda.predict_proba(X)
        combined_probs = (1 - self.lambda_param) * lda_probs + self.lambda_param * qda_probs
        return combined_probs
```
