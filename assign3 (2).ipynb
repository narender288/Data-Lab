# Task_1

```python
df = pd.read_csv('Assignment3.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.iloc[:, 0:5]  
y = df.iloc[:, 5]  
```

```python
X
```

```python
y
```

```python
#  hence we extract the input features.
```

```python
# now 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
y_pred = lr.predict(X)   # predicted , expected output.
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y , y_pred)
```

```python
## we can see that r2_score is very good.
```

```python
sse = mean_squared_error(y, y_pred) * len(y)   # RSS = MSE × n .
print(f" sum of squared error: {sse}")
```
